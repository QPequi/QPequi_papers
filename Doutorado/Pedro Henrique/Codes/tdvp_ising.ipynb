# Quenching TFIM using TDVP method

This code initalizes the transverse field Ising model $H = J\sum_{n=0}^{L-1} \sigma^x_i\sigma^x_{i+1} + g \sum_{n=0}^L \sigma^z_i$ at the all-up state $|\rightarrow\cdots \rightarrow\rangle$ along x-direction.
It then performs a real-time evolution with TDVP and measures magnetizations.
This setup correspond to a global quench from $g =0$ to $g>0$.

```python
tdvp_params = {
        'start_time': 0,
        'dt': dt,
        'trunc_params': {
            'chi_max': D_bond,
            'svd_min': 1.e-10,
            'trunc_cut': None,
        }}
eng = tdvp.TwoSiteTDVPEngine(psi0, M, tdvp_params)
```

```python
def measurement(eng, data):
    keys = ['t', 'Sx', 'Sz', 'trunc_err']
    if data is None:
        data = dict([(k, []) for k in keys])
    data['t'].append(eng.evolved_time)
    data['Sx'].append(eng.psi.expectation_value('Sigmax'))
    data['Sz'].append(eng.psi.expectation_value('Sigmaz'))
    data['trunc_err'].append(eng.trunc_err.eps)
    return data
```

```python
data = measurement(eng, None)
```

```python
while eng.evolved_time < 5.:
    eng.run()
    measurement(eng, data)
```

```python
plt.plot(data['t'], np.sum(data['Sx'], axis=1), label="X")
plt.plot(data['t'], np.sum(data['Sz'], axis=1), label="Z")
plt.plot(data['t'],np.zeros(len(data['t'])), linestyle='--', c='black')

plt.xlabel('time $t$')
plt.ylabel('magnetization')
plt.legend(loc='best')
```
